## Ejercicio 1
Mapeo $x_{n+1}=rx_n(1-x_n)$

```julia
function mapeo2(xi,mui) #defino la función mapeo que va a depender de dos variabes
    lmui=length(mui) #lmui va a ser un vetor de longitud mui y lo mismo con lxi
    lxi=length(xi)
    m=1000
    j=1
    Ax = zeros(lmui,lxi)#hago 2 arreglos, uno para x y otro para mu donde va a ser un arreglo de ceros de la longitud de mu por la de x
    Amu= zeros(lmui,lxi)
    for mu in mui
        xii=xi
        for i in 1:m
        xii = mu*xii.*(1-xii) #La fórmuladel mapeo
        end
        for k in 1:lxi
            Amu[j,k]=mu
        end
        Ax[j,1:end]=xii
        j = j+1
    end
    Ax,Amu
end
```

```julia
xi=linspace(0,1,200)
mui=linspace(0,4,200);
```

```julia
x,y=mapeo2(xi,mui)
plt.scatter(y,x)
```

## Ejercicio 2
Mapeo $x_{n+1}=1-\mu x_n^2$

```julia
#es el mismo procedimiento solo cambia la fórmula.
function mapeo2(xi,mui) 
    lmui=length(mui) 
    lxi=length(xi)
    m=1000
    j=1
    Ax = zeros(lmui,lxi)
    Amu= zeros(lmui,lxi)
    for mu in mui
        xii=xi
        for i in 1:m
        xii = 1-mu*xii.^2 #fórmula del mapeo ()
        end
        for k in 1:lxi
            Amu[j,k]=mu
        end
        Ax[j,1:end]=xii
        j = j+1
    end
    Ax,Amu
    end
```

```julia
xi=linspace(-1,1,200)
mui=linspace(0,2,200);
```

```julia
x,y=mapeo2(xi,mui)
PyPlot.scatter(y,x)
```
